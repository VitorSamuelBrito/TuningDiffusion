```python
import numpy as np
import matplotlib.pyplot  as plt
import math 
```

```python
def compZh(x, dx, dt=1):
    zh = {}
    for val in x:
        val_int = int(val / dx) * dx
        zh[val_int] = zh.get(val_int, 0) + 1
    
    for val in zh:
        zh[val] = float(zh[val]) / dt / dx
    
    return zh

def state(x, x0, x1, dx):
    zheq = compZh(x, dx)
    zhtp = {}
    
    def addZh(zh, ltpx):
        for val in ltpx:
            val_int = int(val / dx) * dx
            # print(val_int)
            zh[val_int] = zh.get(val_int, 0) + 1
            # print(zh)
    
    s = 2  # estado inicial
    ltpx = []
    if x0 > x1:
        x0, x1 = x1, x0  # garante que x0 sempre seja menor que x1
    
    for val in x:
        ltpx.append(val)
        
        if s == 2:
            if val <= x0:
                s = 0
            if val >= x1:
                s = 1    # garante a troca de estados
        
        if val <= x0:
            if s == 1:
                addZh(zhtp, ltpx)
            s = 0
            ltpx = []
        
        if val >= x1:
            if s == 0:
                addZh(zhtp, ltpx)
            s = 1
            ltpx = []
    
    ptpx = []
    for val in zheq:
        ptpx.append((val, float(zhtp.get(val, 0)) / dx / zheq[val]))
    ptpx.sort()
    
    return ptpx
```

```python
x = np.genfromtxt('trajectory_file')#, dtype= None, delimiter= None)

# Para testar a função
x0 = 30
x1 = 37
dx = 0.5 # ideal 0.1 a 0.5 
result = state(x, x0, x1, dx)
hist = compZh(x, dx, dt=1)
# print(result, hist)

ptpx_result = np.asarray(result)
hist_result = []

for i in hist:
    hist_result.append((i, hist[i]))

hist_result = np.asarray(hist_result)

np.savetxt("TPX_fep1d", ptpx_result, fmt="%5.7f")
np.savetxt("hist_fep1d", hist_result, fmt="%5.7f")
```
