```python
import numpy as np 
from itertools import product
import os
```

```python
exaustive = "/home/usuario/Vitor/proj-git/tests_functions/exaustive"

exaustive_C = []
exaustive_W = []
exaustive_H = []
exaustive_x = []

for dtr in sorted(os.listdir(exaustive)): # sorted organiza em ordem alfabetica
    
    if dtr.startswith('results_C'): # separação dos nomes das pastas por vetor
        exaustive_C.append(dtr)
        
    if dtr.startswith('results_W'):
        exaustive_W.append(dtr)
        
    if dtr.startswith('results_H'):
        exaustive_H.append(dtr)
        
    if dtr.startswith('results_x'):
        exaustive_x.append(dtr)
        
# print(exaustive_C) 
# print(exaustive_W)
# print(exaustive_H)
# print(exaustive_x)
```

```python
variavel = "/home/usuario/Vitor/proj-git/tests_functions/teste_variavel"

variavel_C = []
variavel_W = []
variavel_H = []
variavel_x = []

order_var = ['fx', 'fg', 'vx', 'vg']

def init_in(func): # organiza de acordo com a ordem da pasta exaustive
    for i, k in enumerate(order_var):
        if k in func:
            return i
    return len(order_var) 

for dtr in sorted(os.listdir(variavel), key = init_in):
    
    if dtr.startswith('results_C'): # separação dos nomes das pastas por vetor
        variavel_C.append(dtr)
        
    if dtr.startswith('results_W'):
        variavel_W.append(dtr)
        
    if dtr.startswith('results_H'):
        variavel_H.append(dtr)
        
    if dtr.startswith('results_x'):
        variavel_x.append(dtr)
        
        
# print(variavel_C) 
# print(variavel_W)
# print(variavel_H)
# print(variavel_x)
```

```python
C_small = np.array([-20, 18])
W_small = np.array([-13, 17])
H_small = np.array([-7, 8]) # H é o HEIGHT
x_small = np.array([-33, 40])

values1 = list(product(C_small, W_small, H_small)) # combinações sem variavel x
values2 = list(product(C_small, W_small, x_small)) # ## # ## #### ## # H
values3 = list(product(C_small, H_small, x_small)) # ## # ## #### ## # W
values4 = list(product(W_small, H_small, x_small)) # ## # ## #### ## # C

x = np.asarray(values1)
H = np.asarray(values2)
W = np.asarray(values3)
C = np.asarray(values4)

qt = len(C)
```

```python
paths_exa_C = [] # lista que ira receber caminhos dos arquivos 
paths_exa_W = []
paths_exa_H = []
paths_exa_x = []

funcs_exa = ['e24', 'eG', 'grad24', 'gradG']

for k, file in enumerate(exaustive_C):
    for l in range(int(qt)):
        paths_exa_C.append(f"exaustive/{file}/{funcs_exa[k]}_V_{C[l, 0]}_{C[l, 1]}_{C[l, 2]}.dat") 
        # paths recebe o caminho dos arquivos de acordo com as funções e dos vetores
        
for k, file in enumerate(exaustive_W):
    for l in range(int(qt)):
        paths_exa_W.append(f"exaustive/{file}/{funcs_exa[k]}_{W[l, 0]}_V_{W[l, 1]}_{W[l, 2]}.dat")
        
for k, file in enumerate(exaustive_H):
    for l in range(int(qt)):
        paths_exa_H.append(f"exaustive/{file}/{funcs_exa[k]}_{H[l, 0]}_{H[l, 1]}_V_{H[l, 2]}.dat")
        
for k, file in enumerate(exaustive_x):
    for l in range(int(qt)):
        paths_exa_x.append(f"exaustive/{file}/{funcs_exa[k]}_{x[l, 0]}_{x[l, 1]}_{x[l, 2]}_V.dat")
```

```python
paths_var_C = [] 
paths_var_W = []
paths_var_H = []
paths_var_x = []

funcs_var = ['Fx', 'FG', 'Vx', 'VG']

for k, file in enumerate(variavel_C):
    for l in range(int(qt)):
        paths_var_C.append(f"teste_variavel/{file}/{funcs_var[k]}_{C[l, 0]}_{C[l, 1]}_{C[l, 2]}.txt") 
        
for k, file in enumerate(variavel_W):
    for l in range(int(qt)):
        paths_var_W.append(f"teste_variavel/{file}/{funcs_var[k]}_{W[l, 0]}_{W[l, 1]}_{W[l, 2]}.txt")
        
for k, file in enumerate(variavel_H):
    for l in range(int(qt)):
        paths_var_H.append(f"teste_variavel/{file}/{funcs_var[k]}_{H[l, 0]}_{H[l, 1]}_{H[l, 2]}.txt")
        
for k, file in enumerate(variavel_x):
    for l in range(int(qt)):
        paths_var_x.append(f"teste_variavel/{file}/{funcs_var[k]}_{x[l, 0]}_{x[l, 1]}_{x[l, 2]}.txt")
```

```python
## Comparando arquivos do vetor C (M e/ou MAX)

qt2 = len(paths_exa_C)

# listas que amazenarão as somas

var_col = []
var_sum = []

exa_col = []
exa_sum = []

total_sum = []

for comp in range(qt2):
    # leitura dos dados pasta exaustive
    
    exa = np.genfromtxt(f'{paths_exa_C[comp]}')
    exa_col.append(exa[:, 1]) # separação da coluna dois com os results
    exa_sum.append(sum(np.absolute(exa[:, 1]))) # soma da coluna 

    # leitura dos dados pasta test_variavel
    
    var = np.genfromtxt(f'{paths_var_C[comp]}')
    var_col.append(var[:, 1]) 
    var_sum.append(sum(np.absolute(var[:, 1])))

    # diferença entre as somas
    diff = exa_sum[comp] - var_sum[comp]
    total_sum.append(diff)

# salvando resultados comparados
with open('comparative_c.txt', "w") as file:
    for comp in range(qt2):
        file_exa = os.path.basename(paths_exa_C[comp])  # Nome do arquivo exa
        file_var = os.path.basename(paths_var_C[comp])  # Nome do arquivo var

        # escrevendo arquivo_exa - arquivo_var e resultado da soma
        file.write(f"{file_exa} - {file_var}: {total_sum[comp]:.2f}\n")
```
